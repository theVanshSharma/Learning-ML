```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('prices.csv')
data
```

```python
plt.scatter(data.area,data.price)
plt.xlabel('area')
plt.ylabel('price')
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area']],data.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
reg.predict([[5000]])
```

```python
areas = pd.read_csv('areas.csv')
areas.head()
```

```python
prices = reg.predict(areas)
```

```python
areas['prices'] = prices
```

```python
areas
```

```python
areas.to_csv("house_price_prediction")
```
